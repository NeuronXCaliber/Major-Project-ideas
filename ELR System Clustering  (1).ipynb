```python
import numpy as np
import pandas as pd
dataset1=pd.read_csv('studentdatabase.csv')
dataset1.head()
```

```python
from sklearn.cluster import KMeans
x1=np.array([[73,64],[61,61],[61,73],[84,84],[92,67],[43,43],[71,71],[44,44],[45,55],[67,67]])
model1=KMeans(n_clusters=2)
model1.fit(x1)
centroids=model1.cluster_centers_
labels=model1.labels_
```

```python
print(centroids)
```

```python
print(labels)
```

```python
colors=['g.','y.']  ##clustering for student sc % and hsc % on 10 data
for i in range(len(x1)):
    plt.plot(x1[i][0],x1[i][1],colors[labels[i]],markersize=20)
    plt.scatter(centroids[:,0],centroids[:,1],marker='*',s=100)
plt.title('sc % vs hsc %')
plt.show()
```

```python
x2=np.array([[64,32],[61,55],[73,71],[84,89],[64,53],[43,56],[71,55],[44,55],[55,55],[67,55]])
model2=KMeans(n_clusters=2)
model2.fit(x2)
cntrods=model2.cluster_centers_
label2=model2.labels_
print(cntrods)
print(label2)
```

```python
colors=['r.','b.']  ##clustering for student hsc % and ent % on 10 data
for i in range(len(x2)):
    plt.plot(x2[i][0],x2[i][1],colors[label2[i]],markersize=20)
    plt.scatter(cntrods[:,0],cntrods[:,1],marker='*',s=100)
plt.title('hsc % vs ent %')
plt.show()
```
