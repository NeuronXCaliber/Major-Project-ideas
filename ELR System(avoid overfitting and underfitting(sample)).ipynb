```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
dataset1=pd.read_csv('studentdatabase.csv')
dataset1.head()
```

```python
dataset3=pd.read_csv('transactions.csv')
dataset3.head()
```

```python
df1=dataset1.drop(['s_name','age','category'],axis=1)
df1.head()
```

```python
df2=dataset3.drop(['s_id'],axis=1)
df2.head()
```

```python
data=pd.concat([df1,df2],axis=1)
data.head()
```

```python
from sklearn.preprocessing import LabelEncoder
lb=LabelEncoder()
data.iloc[:,-1]=lb.fit_transform(data.iloc[:,-1].values)
data.iloc[:,-3]=lb.fit_transform(data.iloc[:,-3].values)
data.iloc[:,-1]
```

```python
data.head()
```

```python
x=data.iloc[:,:-1]
y=data.iloc[:,-1]
x
```

```python
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
dtree=DecisionTreeClassifier(criterion='entropy',max_depth=3)
dtree.fit(x_train,y_train)
pred1=dtree.predict(x_test)
pred1
```

```python
from sklearn.metrics import accuracy_score
accuracy1=accuracy_score(y_test,pred1)
accuracy1
```

```python
clf=RandomForestClassifier(n_estimators=50,max_depth=10)
clf.fit(x_train,y_train)
pred2=clf.predict(x_test)
pred2
```

```python
accuracy2=accuracy_score(y_test,pred2)
accuracy2
```

```python
def score(model,x,y):
    model.fit(x_train,y_train)
    return model.score(x_test,y_test)
```

```python
score(DecisionTreeClassifier(),x_train,y_train)
```

```python
score(RandomForestClassifier(),x_train,y_train)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
cross_val_score(DecisionTreeClassifier(),x_train,y_train)
```

```python
cross_val_score(RandomForestClassifier(),x_train,y_train)
```

```python
score1=cross_val_score(DecisionTreeClassifier(),x_train,y_train,cv=5)
score1
```

```python
np.average(score1)
```

```python
score2=cross_val_score(RandomForestClassifier(n_estimators=50),x_train,y_train,cv=5)
score2
```

```python
np.average(score2)
```
